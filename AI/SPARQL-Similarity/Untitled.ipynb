```python
import numpy as np
import nltk
import pandas as pd
nltk.download('punkt')

import os
import re
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt


from sklearn.model_selection import train_test_split
```

```python
test = pd.read_json("test-data.json","r", encoding="utf8")
train = pd.read_json("train-data.json","r", encoding="utf8")
```

```python
df = pd.concat([test,train])
```

```python
df.head()
```

```python
def clean_text(text):
    text = text.lower() 
    return text

def cleaner(text):
    text = clean_text(text)
    words = nltk.word_tokenize(text)
    return " ".join(words)

df["question"] = df["corrected_question"].apply(lambda x:  cleaner(x) )
df["query"] = df["sparql_query"].apply(lambda x:  cleaner(x) )
```

```python
df.head()
```

```python
x_train  = df["question"].to_numpy()
y_train_sequences = df["query"].to_numpy()
```

```python
VOCAB_SIZE = 50000
from tensorflow.keras.preprocessing.text import Tokenizer


def vocab_creater(text_lists, VOCAB_SIZE):
    tokenizer = Tokenizer(num_words=VOCAB_SIZE)
    tokenizer.fit_on_texts(text_lists)
    dictionary = tokenizer.word_index

    word2idx = {}
    idx2word = {}
    for k, v in dictionary.items():
        if v < VOCAB_SIZE:
            word2idx[k] = v
            idx2word[v] = k
        if v >= VOCAB_SIZE-1:
            continue
          
    return word2idx, idx2word

word2idx, idx2word = vocab_creater(text_lists=x_train, VOCAB_SIZE=VOCAB_SIZE)

#Redefine .....
VOCAB_SIZE = len(idx2word) +1
```
