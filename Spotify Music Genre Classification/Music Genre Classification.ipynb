```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn import tree
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
import warnings
%matplotlib inline
sns.set(style="ticks", color_codes=True)
warnings.filterwarnings('ignore')
pd.pandas.set_option('display.max_columns',None)
```

```python
fields = ["id", "acousticness", "danceability", "duration_ms", "energy", "instrumentalness",  "key", "liveness", "loudness", "speechiness", "tempo", "valence","genre"]
dataframe = pd.read_csv("whole_dataset_last.csv",sep='\t',encoding='utf-8',names=fields)
dataframe = dataframe.iloc[1:]
```

```python
sns.catplot(x="genre", y="instrumentalness", data=dataframe);
```

```python
def cat_to_number(dataframe,column_name,specify,value):
    dataframe[column_name] = np.where(dataframe[column_name]==specify, value, dataframe[column_name])
    
cat_to_number(dataframe,"genre","metal",0)
cat_to_number(dataframe,"genre","rock",1)
cat_to_number(dataframe,"genre","pop",2)
cat_to_number(dataframe,"genre","classical",3)

x_dataset = dataframe.iloc[:,1:]
y_dataset = dataframe.iloc[:,-1:]

del x_dataset['genre']

print(x_dataset.head())

print(y_dataset.head())
```

```python
for column in x_dataset:
    x_dataset[column] = pd.to_numeric(x_dataset[column])
```

```python
cont_vars = [var for var in x_dataset.columns]
x_dataset[cont_vars].head()
# Let's go ahead and analyse the distributions of these variables

def analyse_continous(df,var):
    df = df.copy()
    df[var].hist(bins=20)
    plt.ylabel('Number of Songs')
    plt.xlabel(var)
    plt.title(var)
    plt.show()

for var in x_dataset.columns: # every numerical and continous variable
    analyse_continous(x_dataset,var)
```

```python
Var_Corr = x_dataset.corr()
# plot the heatmap and annotation on it
sns.heatmap(Var_Corr, xticklabels=Var_Corr.columns, yticklabels=Var_Corr.columns, annot=True)
```

```python
x_dataset['energy'] = x_dataset.energy.astype(float)
x_dataset['loudness'] = x_dataset.loudness.astype(float)

def energy_per_loudness(row):
    return row['loudness'] / row['energy'] 

x_dataset['loudness_per_energy'] = x_dataset.apply(energy_per_loudness, axis=1)

del x_dataset['energy']
del x_dataset['loudness']
del x_dataset['key']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_dataset, y_dataset, test_size = 0.20, random_state = 42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


y_train=y_train.astype('int')
y_test = y_test.astype('int')
```

```python
x_dataset=x_dataset.astype(float)
y_dataset=y_dataset.astype(float)
def run_model(model, alg_name):
    
    print(f"--- {alg_name} ---")
    model.fit(X_train,y_train.values)

    pred = model.predict(X_test)
    print(confusion_matrix(y_test,pred))
    print(f"Accuracy score : {accuracy_score(y_test,pred)}")
    print(classification_report(y_test, pred, [0,1,2,3]))
    scores = cross_val_score(model, x_dataset, y_dataset, cv=5)
    print(np.mean(scores))
```

```python
knn = KNeighborsClassifier(n_neighbors=40)
run_model(knn, "KNN")
```

```python
classifier = RandomForestClassifier(n_estimators = 200, criterion = 'gini',max_depth=8,max_features = 'auto', random_state = 42)
run_model(classifier, "Random Forest")
```

```python
classifier = tree.DecisionTreeClassifier(max_depth = 6,min_samples_leaf=5,random_state=1)
run_model(classifier, "Decision Tree")
```

```python
model = MLPClassifier(alpha= 0.1, hidden_layer_sizes = 11, max_iter = 500, random_state= 0, solver= 'lbfgs')
run_model(model, " MLP Neural network ")
```

```python
logreg=LogisticRegression(C=1000.0,penalty='l1')
run_model(logreg,"Logistic Regression")
```
